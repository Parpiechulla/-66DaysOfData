```python
import pandas as pd
import requests 
import bs4 

url = "https://en.wikipedia.org/wiki/Iron_Chef_America"

website = requests.get(url)

soup = bs4.BeautifulSoup(website.text,"lxml")

right_table = soup.find("table", class_="wikitable sortable")

table_rows = right_table.find_all("tr")

data = []
for row in table_rows:
    data.append([t.text.strip() for t in row.find_all('td')])

df = pd.DataFrame(data,columns=["Iron Chef", "Seasons", "Specialty", "Win", "Loss", "Draw", "Total", "Win%"])
df
```
